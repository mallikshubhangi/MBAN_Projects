```python
list1 = [1,89,87,65,34,98,87,65]
```

```python
def mean(x):
    return sum(x)/len(x) #calculating mean here through sum and len functions

mean(list1)
```

```python
def median(x):
    sorted(x) # works on any sequential object
    if len(x)%2 == 0:
        med = (x[(len(x)//2)]+x[(len(x)//2)-1])/2
    else:
        med = x[len(x)//2]

    return med

median(list1)
```

```python
def mode(x):
    freq_dict = {}
    for i in x:
        if i in freq_dict:
            freq_dict[i] += 1
        else:
            freq_dict[i] = 1

    # Find the mode(s)
    max_freq = max(freq_dict.values())
    modes = [k for k, v in freq_dict.items() if v == max_freq]
    
    return modes

mode(list1)
```

```python
def iqr(x):
    x.sort()
    first_quartile_val = min(x)
    second_quartile_val = x[len(x)//4]
    third_quartile_val = x[(len(x)//4) * 2]
    fourth_quartile_val = x[(len(x)//4) * 3]
    list_quartiles = [first_quartile_val, second_quartile_val, third_quartile_val, fourth_quartile_val]

    return list_quartiles

iqr(list1)
```

```python
iqr_range = iqr(list1)[-1] - iqr(list1)[1] #this is the actualy iqr range, the code above gives the iqrs
print(iqr_range)
```

Also, with regard to this. You would use Encoding for any categorical column. You do this because most models will not understand “text” or string values properly. So to help them with “understanding” these values you would use one form of encoding. However, there are two types. For example, if there was a column called “Difficulty” and the values were – Low, Medium and High, you would use Label Encoding (by using the map function in Python), to convert Low to 1, Medium to 2 and High to 3, that is because there is an order to these values ( 3 > 2 > 1) However, if the column was something like “Colour”, and the values were Red, Blue and Green, doing this would not make any sense, because Red is not “greater” or “bigger” or “lesser” than the other colours. So to help the model understand this better, you would create dummy variables (one hot encoding). So from one column Colour, there would three new columns created – Colour_Red, Colour_Green and Colour_Blue. And if the colour of the row was Red originally, the values for that row would now be 1 in Colour_Red and 0 in Colour_Green and Colour_Blue. If the colour was Green, then the values would be 0 in Colour_Red and Colour_Blue, but 1 in Colour_Green.
